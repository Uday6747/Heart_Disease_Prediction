```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import logging,sys
```

```python
df=pd.read_csv("C:\\Users\\91903\\Desktop\\ML\\breast Cancer\\dataset.csv",header=1)
dataset = shuffle(np.array(df))
```

```python
df.head()
```

```python
df.drop(df.columns[[0]], axis=1, inplace=True)
dataset = shuffle(np.array(df))
```

```python
df.head()
```

```python
extracted_dataset= []
target = []

#extract target column
for row in dataset:
    extracted_dataset.append(row[1:])
    target.append(row[0])

```

```python
target
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(extracted_dataset,target,test_size=0.3)
```

```python
model = MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)
model.fit(X_train,Y_train)
```

```python
predicted = model.predict(X_test)
predicted
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predicted,normalize=True)
```
